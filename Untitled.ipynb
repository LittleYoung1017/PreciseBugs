```python
def getProject(cve):
    url='https://nvd.nist.gov/vuln/detail/'+cve

    response = requests.get(url)
    print(response)
    for line in str(response.text).split('\n'):
        if 'github.com/' in line:
            projectName=line.split('github.com/')[1]
            if '/' in projectName:
                projectName=projectName.split('/')[1]
            print(projectName)
            return projectName
        


    print(response)
```
